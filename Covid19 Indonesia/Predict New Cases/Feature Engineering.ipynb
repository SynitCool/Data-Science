```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df.head(17)

# Growth Factor of New Cases is calculate by present new cases/past new cases
# Growth Factor of New Deaths is calculate by present new deaths/past new deaths
# And the value is missing because it divides by zero
```

```python
df["Growth Factor of New Cases"].fillna(0, inplace=True)
df["Growth Factor of New Deaths"].fillna(0, inplace=True)

df.head(17)
```

```python
df.info()
```

```python
features = ["New Cases","New Deaths", "New Recovered", "New Active Cases"]

fig, ax = plt.subplots(figsize=(15, 9), nrows=2, ncols=2)

for i in range(2):
    sns.boxplot(data=df, x=df[features[0:2][i]], ax=ax[i][0])
    
    ax[i][0].set_xlabel("")
    ax[i][0].set_title(f"Boxplot of {features[0:2][i]} feature")
            
    sns.boxplot(data=df, x=df[features[2:4][i]], ax=ax[i][1])
            
    ax[i][1].set_xlabel("")
    ax[i][1].set_title(f"Boxplot of {features[2:4][i]} feature")
            
plt.show()
```
