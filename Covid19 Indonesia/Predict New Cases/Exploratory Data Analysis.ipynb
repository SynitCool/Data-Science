```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Dataset\\covid19ind.csv")

df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
features = ["Longitude", "Latitude", "Area (km2)", 
            "Population", "Population Density", "Total Cities",
            "Time Zone", "Island", "Province", "City or Regency",
            "Total Regencies", "Total Districts", "Special Status",
            "Location Level", "Country", "Continent", "Total Urban Villages",
            "Total Rural Villages", "Location", "Location ISO Code"]

for feature in features:
    df.drop(columns=[feature], inplace=True)
```

```python
df.head()
```

```python
df.info()
```

```python
features = ["Case Fatality Rate", "Case Recovered Rate"]


for feature in features:
    df[feature] = df[feature].str.split(pat="%", expand=True)[0]
    
    df[feature] = df[feature].astype(float)
```

```python
df.info()
```

```python
df.describe()
```

```python
features_1 = ["New Cases", "New Deaths"]
features_2 = ["New Recovered", "New Active Cases"]

fig, ax = plt.subplots(figsize=(15, 9), ncols=2, nrows=2)

for i in range(2):
    sns.distplot(x=df[features_1[i]], ax=ax[i][0])
    
    ax[i][0].set_xlabel("")
    ax[i][0].set_title(f"Distribution of {features_1[i]} feature")
    
    sns.distplot(x=df[features_2[i]], ax=ax[i][1])
    
    ax[i][1].set_xlabel("")
    ax[i][1].set_title(f"Distribution of {features_2[i]} feature")
    
plt.show()
```

```python
features = ["Total Cases", "Total Deaths", "Total Recovered", "Total Active Cases"]

fig, ax = plt.subplots(figsize=(15, 9), nrows=2, ncols=2)

for i in range(2):
    sns.distplot(x=df[features[0:2][i]], ax=ax[i][0])
    
    ax[i][0].set_xlabel("")
    ax[i][0].set_title(f"Distribution of {features[0:2][i]} feature")
    
    sns.distplot(x=df[features[2:4][i]], ax=ax[i][1])
    
    ax[i][1].set_xlabel("")
    ax[i][0].set_title(f"Distribution of {features[2:4][i]} feature")
    
plt.show()
```

```python
features = ["New Cases per Million", "Total Cases per Million", "New Deaths per Million", "Total Deaths per Million"] 

fig, ax = plt.subplots(figsize=(15, 9), nrows=2, ncols=2)

for i in range(2):
    sns.distplot(x=df[features[0:2][i]], ax=ax[i][0])
    
    ax[i][0].set_xlabel("")
    ax[i][0].set_title(f"Distribution of {features[0:2][i]} feature")
    
    sns.distplot(x=df[features[2:4][i]], ax=ax[i][1])
    
    ax[i][1].set_xlabel("")
    ax[i][1].set_title(f"Distribution of {features[2:4][i]} feature")
    
plt.show()
```

```python
features = ["Case Fatality Rate", "Case Recovered Rate", "Growth Factor of New Cases", "Growth Factor of New Deaths"]

fig, ax = plt.subplots(figsize=(15, 9), nrows=2, ncols=2)

for i in range(2):
    sns.distplot(x=df[features[0:2][i]], ax=ax[i][0])
    
    ax[i][0].set_xlabel("")
    ax[i][0].set_title(f"Distribution of {features[0:2][i]} feature")
    
    sns.distplot(x=df[features[2:4][i]], ax=ax[i][1])
    
    ax[i][1].set_xlabel("")
    ax[i][1].set_title(f"Distribution of {features[2:4][i]} feature")
    
plt.show()
```
