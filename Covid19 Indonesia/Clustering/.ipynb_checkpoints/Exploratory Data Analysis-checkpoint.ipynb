```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Dataset\\Covid19propinsi.csv")

df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df = df.rename(columns={"Active_Case":"Daily_Active_Case"})

df.head()
```

```python
daily_features = []

for column in df.columns:
    if "Daily_" in column:
        daily_features.append(column)
        
fig, ax = plt.subplots(figsize=(15, 9), ncols=2, nrows=2)

for i in range(2):
    sns.distplot(x=df[daily_features[:2][i]], ax=ax[i][0])
    
    ax[i][0].set_xlabel("")
    ax[i][0].set_title(f"Distribution of {daily_features[:2][i]} feature")
    
    sns.distplot(x=df[daily_features[2:][i]], ax=ax[i][1])
    
    ax[i][1].set_xlabel("")
    ax[i][1].set_title(f"Distribution of {daily_features[2:][i]} feature")
    
plt.show()
```

```python
cumulative_features = []

for column in df.columns:
    if "Cumulative_" in column:
        cumulative_features.append(column)
        
fig, ax = plt.subplots(figsize=(15, 9), nrows=2, ncols=2)

for i in range(2):
    sns.distplot(x=df[cumulative_features[:2][i]], ax=ax[i][0])
    
    ax[i][0].set_xlabel("")
    ax[i][0].set_title(f"Distribution of {cumulative_features[:2][i]} feature")
    
    sns.distplot(x=df[cumulative_features[2:][i]], ax=ax[i][1])
    
    ax[i][1].set_xlabel("")
    ax[i][1].set_title(f"Distribution of {cumulative_features[2:][i]} feature")

plt.show()
```

```python
top_10_daily_case = pd.DataFrame(data={"Province":df.iloc[df["Daily_Case"].sort_values(ascending=False)[:10].index]["Province"],
                                       "Daily_Case":df.iloc[df["Daily_Case"].sort_values(ascending=False)[:10].index]["Daily_Case"]})

fig, ax = plt.subplots(figsize=(20, 9))

sns.barplot(data=top_10_daily_case, x="Province", y="Daily_Case")
plt.title("Plotting of Top 10 Daily Case")

plt.show()
```
